```python
balance = 0

while True:
    print(''' Menu
    1. Check Balance
    2. Credit Amount
    3. Debit Amount
    4. Exit''')
    
    try:
        choice  = int(input("enter a choice(1-4): "))
    except ValueError:
        print("Invalid input. Please enter numbers only.\n")
        continue
   
    if choice == 1:
        print(f'Available Balance: ₹{balance:,.2f}\n')
        
    elif choice == 2:
        try:
            amount = float(input("Enter the Amount to be credited: "))
        except ValueError:
            print("Invalid input. Please enter numbers only.\n")
            continue
            
        if amount  <= 0:
            print("invalid amount\n")
        else:
            balance += amount
            print(f"Amount credited successfully. New balance: ₹{balance:,.2f}\n")
            
    elif choice == 3:
        try:
            amount = float(input("Enter the Amount to be debited: "))
        except ValueError:
            print("Invalid input. Please enter numbers only.\n")
            continue
            
        if amount  <= 0:
            print("invalid amount\n")
        elif amount > balance:
             print("low balance\n")
        else:
            balance -= amount
            print(f'''₹{amount} debited successfully
                      current balance: ₹{balance:,.2f}\n''')
            
    elif choice == 4:
        print("Exiting application...")
        break
    else:
        print("invalid Choice\n")

print("exited successfully")
```
