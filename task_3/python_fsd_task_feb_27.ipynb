<h1>Make a bank with 2 types of accounts </h1>
<h3>
i) savings account<br>
ii) Current account<br>
The current account has an overdraft limit, which is like a credit amount.<br>
Think if you don't have an amount in your current account, but still you can make a transaction till the overdraft limit is reached, give a different overdraft limit for different accounts<br>
<br>
The key ideology is Inheritance
Withdrawal and Deposit can be in the parent class.</h3>

```python
class bank_account:

    def __init__(self, account_holder_name, account_balance):
        self.name = account_holder_name
        self.balance = account_balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print("amount deposited successfully")
            
        else:
            print("invalid amount")

    def withdraw(self,amount):
        if amount <= 0:
            print("invalid amount")
        elif amount > self.balance:
            print("low balance")
        else:
            self.balance-= amount
            print("withdraw successfully")
```

```python
#savings account
class savings_account(bank_account):
    pass
```

```python
#current account
class current_account(bank_account):
    def __init__(self, account_holder_name, account_balance, account_overdraft_limit):
        super().__init__(account_holder_name, account_balance)
        self.overdraft_limit = account_overdraft_limit

    def withdraw(self, amount):
        if amount <= 0: 
            print("amount invalid")
        elif amount > self.balance + self.overdraft_limit:
            print("overdraft limit exceeded")
        elif amount <= self.balance + self.overdraft_limit:
            self.balance -= amount
            print("successfully withdrawn")
```
