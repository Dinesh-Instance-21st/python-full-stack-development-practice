<h2>1. Find odd or even

```python
num=int(input("Enter a number to check whether its even or odd:"))

def check_odd_even(num):
    return "even" if num%2 == 0 else "odd"
    
print(check_odd_even(num))
```

<h2>2. sum of n numbers

```python
num = int(input("Enter a number:"))
total_sum=0
for i in range (num+1):
    total_sum+=i
print(total_sum)
```

<h2>Question 4: Student Marks Dictionary

```python
#Write a Python program that creates a dictionary with student names as keys and their marks as values.
student_marks = {}
student_marks["dinesh"] = 40
student_marks["harish"] = 50
student_marks["vicky"] = 60
student_marks["kishore"] = 50
student_marks["pallava"] = 80

#Print the name of the student who scored the highest marks.
max_mark=0
top_student=""

for name, mark in student_marks.items():
    if mark > max_mark:
        max_mark = mark
        top_student=name

print(f"Top Scorer: {top_student}")

#Add a new student with marks to the dictionary.
student_marks["arun"]=1

#Print the updated dictionary.
print(student_marks)
```

Question 5: Remove Duplicates using Set</h2>

```python
#Write a Python program that creates a list containing duplicate values.

nums = []

for i in range(10):
    nums.append(i)
    nums.append(7)
    nums.append(3)


print(nums)

#Convert the list into a set.
unique_nums =set(nums)

#Print the unique values.
print(unique_nums)
```

<h2>
Question 8: Simple Data Analysis</h2>

```python
#Write a Python program that creates a list of 10 numbers.
import random
nums = [random.randint(1,100) for _ in range(10)]

print(nums)
```

```python
# Write three separate functions:

# One function to calculate the average.

def get_avg():
    return sum(nums)/len(nums)
    
# One function to find the maximum value.

def get_max_value():
    return max(nums)

# One function to find the minimum value.
def get_min_value():
    return min(nums)

# Call all three functions and print the results clearly.
print(f'''average: {get_avg()}
maximum value: {get_max_value()}
minimum value: {get_min_value()}''')
```
